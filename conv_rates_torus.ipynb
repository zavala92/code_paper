# Convergence rates for torus with radii R=2, r=1

```python
import matplotlib.pyplot as plt
import numpy as np
x2=np.array([5.537647e-03,3.781855e-04,2.412797e-05,1.515720e-06, 9.485241e-08]); 
x3=np.array([1.046193e-03,6.997348e-05,4.471644e-06,2.810634e-07,1.759229e-08]);
x4=np.array([7.104625e-05,1.199858e-06,1.904660e-08,2.992669e-10,4.870595e-12]);
x5=np.array([2.066624e-05,3.192766e-07,5.083293e-09,7.655854e-11,1.134529e-12]);
h=np.array([8.115537e-01,4.057768e-01,2.028884e-01,1.014442e-01,5.072211e-02]);
```

Convergence rates for $k=0\, mod(2)$

```python
plt.subplots(figsize=(8, 6));
plt.loglog(h, x2, '-*b', label="$\Vert \int_{\mathcal{M}}fdS-\int_{\mathcal{M}_{h}^{k}}f_{k}dS\Vert_{\mathbb{L}\infty},k=2$")
plt.loglog(h, x4, '-*r', label="$\Vert \int_{\mathcal{M}}fdS-\int_{\mathcal{M}_{h}^{k}}f_{k}dS\Vert_{\mathbb{L}\infty},k=4$")
plt.loglog(h, np.power(h,2+2)*x2[0]/np.power(h[0],2+2), '--b',label="$h^{4}$")
plt.loglog(h, np.power(h,4+2)*x4[0]/np.power(h[0],4+1), '--r', label="$h^{6}$")
plt.xlabel("$h$",fontsize=12, color="black");
plt.ylabel("error", fontsize=12, color="black");
plt.legend()
plt.savefig('error_plot_k=even.pdf') 
```

Convergence rates for $k=1\, mod(2)$

```python
plt.subplots(figsize=(8, 6));
plt.loglog(h, x3, '-*b', label="$\Vert \int_{\mathcal{M}}fdS-\int_{\mathcal{M}_{h}^{k}}f_{k}dS\Vert_{\mathbb{L}\infty},k=3$")
plt.loglog(h, x5, '-*r', label="$\Vert \int_{\mathcal{M}}fdS-\int_{\mathcal{M}_{h}^{k}}f_{k}dS\Vert_{\mathbb{L}\infty},k=5$")
plt.loglog(h, np.power(h,3+1)*x3[0]/np.power(h[0],3+1), '--b', label = "$h^{4}$")
plt.loglog(h, np.power(h,5+1)*x5[0]/np.power(h[0],5+1), '--r', label = "$h^{6}$")
plt.xlabel("$h$",fontsize=12, color="black");
plt.ylabel("error", fontsize=12, color="black");
plt.legend()
plt.savefig('error_plot_k=odd.pdf') 
```
